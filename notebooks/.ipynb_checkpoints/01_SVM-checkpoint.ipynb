```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1.0 - x

def cls(x,y):
    return y >= f(x)
```

```python
X = np.random.rand(1000,2)
y = [cls(*x) for x in X]
```

```python
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.plot([x[0] for x in X if cls(*x)], [x[1] for x in X if cls(*x)], 'r.')
plt.plot([x[0] for x in X if not cls(*x)], [x[1] for x in X if not cls(*x)], 'bx')

px = np.linspace(0, 1, 10)
plt.plot(px, f(px), 'g-')
plt.show()
```

```python
import sklearn
from sklearn import svm

clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
X_test = np.random.rand(100,2)
y_test = [cls(*x) for x in X_test]
```

```python
y_pred = clf.predict(X_test)
```

```python
sklearn.metrics.accuracy_score(y_test, y_pred)
```

```python
clf.score(X_test, y_test)
```

```python
def f_pred(x):
    return (-clf.intercept_[0] - clf.coef_[0][0] * x) / (clf.coef_[0][1])

px = np.linspace(0, 1, 10)
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.plot([x[0] for x in X_test if cls(*x)], [x[1] for x in X_test if cls(*x)], 'r.')
plt.plot([x[0] for x in X_test if not cls(*x)], [x[1] for x in X_test if not cls(*x)], 'bx')

plt.plot(px, f_pred(px), 'r-')
plt.plot(px, f(px), 'g-')
plt.show()
```
